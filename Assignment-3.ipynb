### Using Selenium

```python
from selenium import webdriver
import requests
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
def get_books_links(pageno):
    url = "https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_"+str(pageno)+"?ie=UTF8&pg="+str(pageno)
    # print(url)
    driver.get(url)
    
    grids = driver.find_elements(By.CLASS_NAME, 'zg-grid-general-faceout')
    lst = []
    for el in grids:
        link = el.find_element(By.TAG_NAME, 'a')
        lst.append(link.get_attribute('href'))
    return lst
```

```python
def get_books_details(urls):
    complete = []
    for url in urls:
        # print(url)
        driver.get(url)
        sub = []
        book_name = driver.find_element(By.ID, 'productTitle')
        sub.append(book_name.text)
        
        #getting unordered list
        unlst = driver.find_element(By.ID, 'detailBulletsWrapper_feature_div')
        lstitems = unlst.find_elements(By.TAG_NAME, 'li')
        count = 0
        # for item in lstitems:
        #     if(count<7):
        #         # print(item.text)
        #         res = item.text.partition(': ')[2]
        #         sub.append(res)
        #     count = count + 1
        # complete.append(sub)
        
        f = 0
        for item in lstitems:
            if "Publisher" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
            
        f = 0
        for item in lstitems:
            if "Language" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
        
        f = 0
        for item in lstitems:
            if "Paperback" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
            
        f = 0
        for item in lstitems:
            if "ISBN-10" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
            
        f = 0
        for item in lstitems:
            if "ISBN-13" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
            
        f = 0
        for item in lstitems:
            if "Weight" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
            
        f = 0
        for item in lstitems:
            if "Dimensions" in item.text:
                # print(item.text)
                res = item.text.partition(': ')[2]
                sub.append(res)
                f=1
                break
        if f==0:
            sub.append('NaN')
        
        
        hdr = driver.find_element(By.ID, 'cm-cr-dp-review-header')
        if "No customer reviews" not in hdr.text:
            reviews = driver.find_element(By.ID, 'cm-cr-dp-review-list')
            sub.append(reviews.text.replace("\n"," "))
        else:
            sub.append('NaN')
  
        complete.append(sub)
    return complete
```

Q2: Consider the following Link:
https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page=1

The page contains information about movies released in the year 2020 sorted according to number
of votes. Scrape the following information for the first 1000 movies.
- Movie name
- IMDB Rating
- Description 
- Votes
- Director(s)
- Stars 
- Metascore
- Movie Genre
- Total Duration

### Using selenium

```python
from selenium import webdriver
import requests
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
def imdb_data(pageno):
    #url = "https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page=1"
    url = "https://www.imdb.com/search/title/?release_date=2020-01-01,2020-12-31&sort=num_votes,desc&start="+str(50*(pageno-1)+1)+"&ref_=adv_nxt"
    driver.get(url)
    complete = [] 
 
    titles = driver.find_elements(By.CLASS_NAME, "lister-item-header") 
    sub=[]
    for title in titles:
        sub.append(title.text[3:len(title.text)-7])
    complete.append(sub)

    imdb = driver.find_elements(By.TAG_NAME, 'strong')
    sub2=[]
    for rating in imdb:
        # print(rating.text)
        sub2.append(rating.text)
    complete.append(sub2[2:])  
    
    desc = [driver.find_elements(By.CLASS_NAME, 'text-muted')]
    # print(desc)
    sub3=[]
    for d in desc[0]:
        if len(d.text) > 55:
            sub3.append(d.text)
            # print(d.text)
            # print("\n")
    complete.append(sub3)
    
    
    votes = [driver.find_elements(By.CLASS_NAME, 'sort-num_votes-visible')]
    # print(votes)
    sub4=[]
    for v in votes[0]:
        #print(v.text[7:14])
        sub4.append(v.text[7:14])
    complete.append(sub4)
    
    meta_scores = driver.find_elements(By.CLASS_NAME, 'ratings-bar')
    # print(meta_scores)
    sub6=[]
    for meta_score in meta_scores:
        if(meta_score.text[14:16] != ""):
            sub6.append(meta_score.text[14:16])
        else:
            sub6.append('NaN')
        # print(meta_score.text[13:16])
    complete.append(sub6)
        
    genre = driver.find_elements(By.CLASS_NAME, 'genre')
    sub5=[]
    for g in genre:
        sub5.append(g.text)
    complete.append(sub5) 
    
    durations = driver.find_elements(By.CLASS_NAME, 'runtime')
    sub7=[]
    for duration in durations:
        sub7.append(duration.text)
        # print(duration.text)
    complete.append(sub7)
    
    return complete
```

### Using beautifulSoup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
def imdb_data(pageno):
    #url = "https://www.imdb.com/search/title/?release_date=2020&sort=num_votes,desc&page=1"
    url = "https://www.imdb.com/search/title/?release_date=2020-01-01,2020-12-31&sort=num_votes,desc&start="+str(50*(pageno-1)+1)+"&ref_=adv_nxt"
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser') 
    complete = [] 
    for d in soup.find_all('div', attrs={'class' : "lister-item mode-advanced"}):
        head = d.find('h3', attrs={'class':"lister-item-header"})#.replace('\n','')
        movie_name = head.find('a')

        imdb_div = d.find('div', attrs={'class':"inline-block ratings-imdb-rating"})
        imdb_rating = imdb_div.find('strong')
        
        desc_div = d.find('div', attrs={'class':"lister-item-content"})
        desc = desc_div.find_all('p', attrs={'class':"text-muted"})[1:2]
        
        votes_p = d.find('p', attrs={'class':"sort-num_votes-visible"})
        votes = votes_p.find('span', attrs={'name':'nv'})
        
        direc_ = d.find('p', attrs={'class':""})
        direc=[]
        
        check = direc_.find('span')
        
        for name in direc_.find_all(['a', 'span']):
            direc.append(name)
            
        direc_final=[]
        stars_final=[]
        
        if check is not None:
            for el in direc:
                direc_final.append(el)
                if el == check :
                    break
            
            for el in direc:
                if(el not in direc_final):
                    stars_final.append(el)
            
            direc_final.pop()
            
        else:
            stars_final = direc
            
        meta_score = d.find('div', attrs={'class':"inline-block ratings-metascore"})
        genre = d.find('span', attrs={'class':"genre"})
        duration = d.find('span', attrs={'class':"runtime"})
            
        sub = []
                        
        if movie_name is not None: # dealing with missing data
            sub.append(movie_name.text)
        else:
            sub.append('NaN')
        if imdb_rating is not None: 
            sub.append(imdb_rating.text)
        else:
            sub.append('NaN')
        if desc is not None: 
            sub.append(desc[0].text.strip())
        else:
            sub.append('NaN')
        if votes is not None: 
            sub.append(votes.text)
        else:
            sub.append('NaN')
        
        if direc_final != []: 
            sub.append((np.array(direc_final)).flatten()) 
        else:
            sub.append('NaN')
        
        if stars_final != []:
            sub.append((np.array(stars_final)).flatten())
        else:
            sub.append('NaN')

        if meta_score is not None: 
            sub.append(meta_score.text.replace("\n","")[:3]) 
        else:
            sub.append('NaN')
        if genre is not None: 
            sub.append(genre.text.strip())
        else:
            sub.append('NaN')
        if duration is not None: 
            sub.append(duration.text)
        else:
            sub.append('NaN')
        
        complete.append(sub)
        
    return complete   
```
