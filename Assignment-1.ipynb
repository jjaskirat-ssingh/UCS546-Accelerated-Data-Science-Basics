#### (d) Find the most frequent value and their indice(s) in the following arrays: 
#### i. x = np.array([1,2,3,4,5,1,2,1,1,1]) 
#### ii. y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ]) 

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
el = np.bincount(x).argmax()
print("Most frequent element: ",el,"\n")
print("Indices at which ", el," occurs:")
for i in range(x.size):
    if(x[i] == el):
        print(i)
```

```python
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3, ])
el = np.bincount(y).argmax()
print("Most frequent element: ",el,"\n")
print("Indices at which ", el," occurs:")
for i in range(y.size):
    if(y[i] == el):
        print(i)
```

#### (f) For the matrix: n_array = np.array([[55, 25, 15],[30, 44, 2],[11, 45, 77]]), find

#### (g) Multiply the following matrices and also find covariance between matrices using NumPy: 
#### i. p = [[1, 2], [2, 3]] ; q = [[4, 5], [6, 7]] 
#### ii. p = [[1, 2], [2, 3], [4, 5]] ; q = [[4, 5, 1], [6, 7, 2]]


```python
p = np.array([[1,2],[2,3]])
q = np.array([[4,5],[6,7]])
print("Dot Product: \n", np.dot(p, q))
print("\nCovariance Matrix of the 2 matrices: \n", np.cov(p,q))
```

```python
p = np.array([[1, 2], [2, 3], [4, 5]])
q = np.array([[4, 5, 1], [6, 7, 2]])
print("Dot Product: \n", np.dot(p, q))
q = np.reshape(q, (3,2))
print("\nCovariance Matrix of the 2 matrices: \n", np.cov(p,q)) # gives ValueError without reshaping 
```

#### (a) Write a function named img_to_array(path) that reads an image from a specified path and save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale images) 

```python
from PIL import Image
from IPython.display import display
import numpy as np

def img_to_arr(path):
#     image = path
#     text = pytesseract.image_to_string(Image.open(image), lang="eng")
#     print(text)

    img = np.array(Image.open(path))
    print(type(img))
    print(img.dtype)
    print(img.shape)
    
    #print(np.array(Image.open(path)))
    imm = Image.open(path)
    

    if(imm.mode == 'RGB'): # for RGB image
        print("\n", imm.mode)
        print("It is an RGB image.")
        display(imm)
        
        file = open("rgbimage.txt", "w+")
        content = str(img)
        file.write(content)
        file.close()

    elif(imm.mode == 'L'): # for greyscale image
        print("\n", imm.mode)
        print("It is a Greyscale image.")
        display(imm)
        file = open("greyscale.txt", "w+")
        content = str(img)
        file.write(content)
        file.close()
    
    
path = './jas.jpeg' 
path2 = './greyscale.jpeg'
img_gray = Image.open(path).convert('L')
img_gray.save(path2)

img_to_arr(path) #RGB
img_to_arr(path2) #Greyscale
```
