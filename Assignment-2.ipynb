### Q3: Based on Constructing Term Document Matrix 

#### For the preprocessed corpus, construct a Term Document Matrix (using both inbuilt methods and from scratch), whose entries are: 

#### (a) Binary

```python
# Binary TDM from scratch: 
content = final
reviews = [i for i in range(1, len(corpus)+1)]
rev_count={} #dictionary for each review
for i in range(len(content)):
    word_count={} #dictionary for words for each review
    for word in content[i].split():
        if word not in word_count.keys():
            word_count[word]=1
    rev_count[reviews[i]]=word_count
# rev_count
```

```python
binary_tdm_scratch=pd.DataFrame(rev_count)
binary_tdm_scratch.fillna(0,inplace=True)
binary_tdm_scratch
```

```python
# Binary TDM using inbuilt function: 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=True)
X=cv.fit_transform(content) # returns a binary tdm in compressed sparse row format
# print(X)
```

```python
binary_tdm_inbuilt=pd.DataFrame(X.toarray().T,index=cv.get_feature_names(),columns=reviews)
binary_tdm_inbuilt
```

#### (b) Actual Term Frequency 

```python
# Term frequency TDM from scratch
rev_count1={} #dictionary for each review
for i in range(len(content)):
    word_count1={} #dictionary for words for each review
    for word in content[i].split():
        if word not in word_count1.keys():
            word_count1[word]=0
        word_count1[word]+=1
    rev_count1[reviews[i]]=word_count1
rev_count1
```

```python
#Converting dictionary to term frequency TDM dataframe
tdm_tf_scratch=pd.DataFrame(rev_count1)
tdm_tf_scratch.fillna(0,inplace=True)
tdm_tf_scratch
```

```python
# Term frequency TDM using inbuilt function: CountVectorizer of Sklearn library
from sklearn.feature_extraction.text import CountVectorizer
cv1=CountVectorizer()
Y=cv1.fit_transform(content) # returns a term frequency tdm in compressed sparse row (csr) format 
# print(Y)
```

```python
# Converting csr format to dataframe
tdm_tf_inbuilt=pd.DataFrame(Y.toarray().T,index=cv1.get_feature_names(),columns=reviews)
tdm_tf_inbuilt
```

#### (c) Term Frequency with length normalization 

```python
# Term Frequency with length Normalization TDM from scratch
tdm_length_norm_scratch=tdm_tf_scratch.iloc[:,:].div(tdm_tf_scratch.sum(axis=0),axis=1)
tdm_length_norm_scratch
```

```python
#Term Frequency with length Normalization TDM using TfIdfVectorizer class of sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer(use_idf=False,norm='l1')
Z=tf.fit_transform(content) # returns length normalized csr
```

```python
#Converting CSR format to dataframe
tdm_len_norm_inbuilt=pd.DataFrame(Z.toarray().T,index=tf.get_feature_names(),columns=reviews)
tdm_len_norm_inbuilt
```

#### (d) Term Frequency-Inverse Document Frequency 

```python
import numpy as np
```

```python
#Computing idf from scratch
no_of_revs=len(reviews)
unique_terms=tdm_length_norm_scratch.index #as idf is computed for each unique term of corpus
idf_count={} #to store idf score of each term of corpus
for term in unique_terms:
    idf_count[term]=0 #initializing entry in idf dictionary
    for doc in content:
        if term in doc.split():
            idf_count[term]+=1
    idf_count[term]=np.log10(no_of_revs/idf_count[term])
idf_count
```

```python
#Converting idf_count dictionary to DataFrame
idf_series=pd.Series(idf_count)
idf=pd.DataFrame(idf_series)
idf
```

```python
#Computing tf-idf by multiplying tf and idf
tf_idf_scratch=tdm_length_norm_scratch*idf.values
tf_idf_scratch
```

```python
#Term Frequency-Inverse Document Frequency TDM using TfIdfVectorizer class of sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
tf1=TfidfVectorizer()
U=tf1.fit_transform(content)
tf_idf_inbuilt=pd.DataFrame(U.toarray().T,index=tf1.get_feature_names(),columns=reviews)
tf_idf_inbuilt
```

### Q4: Based on Co-occurrence Matrix

#### (a) Construct a term-term co-occurrence matrix from the pre-processed list of documents (obtained from Q2) whose each (ij)th entry is number of documents in which both ith and jth terms co-occur. 

```python
# We have to use list 'final' from Q2
final
```

```python
reviews = [i for i in range(1, len(corpus)+1)]
rev_count={}
for i in range(len(final)):
    word_count={} 
    for word in final[i].split():
        if word not in word_count.keys():
            word_count[word]=1
    rev_count[reviews[i]]=word_count
rev_count
mat=pd.DataFrame(rev_count) # mat is a binary tdm matrix
mat.fillna(0,inplace=True)
# mat
```

```python
co_occur_matrix=np.dot(mat.values,mat.values.T)
np.fill_diagonal(co_occur_matrix,0)
co_occur_matrix=pd.DataFrame(co_occur_matrix, index=mat.index, columns=mat.index)
co_occur_matrix
```

```python
# import math
def pa(i,j):
    sum=0
    for rev in final:
        if i in rev and j in rev: # if both words are in the review
            sum+=1
    return sum

def qa(i):
    sum=0
    for rev in final: 
        if i in rev: # if word in review
            sum+=1
    return sum
```
